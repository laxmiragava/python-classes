```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
y = np.array([2,7,1,9])
plt.plot(y,marker='D',ms=10,mec= 'r',mfc = 'y')
plt.xlabel('x axis points')
plt.ylabel('y axis points')
plt.title('marker plot')
plt.show()
```

```python
y = np.array([2,7,1,5,3])
plt.plot(y,marker='h',ms=10,mec= 'm',mfc = 'b')
plt.xlabel('x axis points')
plt.ylabel('y axis points')
plt.title('marker plot')
plt.show()
```

```python
y = np.array([2,7,1,5,3])
plt.plot(y,linewidth ='10', marker='D',ms=10,mec= 'm',mfc = 'r')
plt.xlabel('x axis points')
plt.ylabel('y axis points')
plt.title('marker plot')
plt.show()
```

#     BAR PLOT

```python
langs = [ 'c' , 'c++' , 'java' , 'python' , 'php']
students = [21 , 16 ,35 ,27 ,12]
plt.bar(langs,students)
plt.show()
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25,],[50,40,70,80,20],label='BMW',width = .5)
plt.bar([0.75,1.75,2.75,3.75,4.75,],[80,20,20,50,70],label='audi',color='r',width = .5)
plt.legend()
plt.xlabel('NO OF DAYS')
plt.ylabel('DISTENCE IN KMS')
plt.title('MAILEGE OFBMW AND AUDI ')
plt.show()
```
