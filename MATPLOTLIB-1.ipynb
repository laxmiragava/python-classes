```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = [5,2,7]
y = [2,16,4]
plt.plot(x,y)
plt.xlabel('x axis points')
plt.ylabel('y axis points')
plt.title('narmal plot')
plt.show()
```

```python
x = np.array ([0,6])
y = np.array ([0,9])
plt.plot(x,y)
plt.xlabel('x axis points')
plt.ylabel('y axis points')
plt.title('narmal plot')
plt.show()
```

```python
x = np.array ([1,8])
y = np.array ([3,10])
plt.plot(x,y,'o')
plt.xlabel('x axis points')
plt.ylabel('y axis points')
plt.title('narmal plot')
plt.show()
```

```python
x = np.array ([1,2,6,8])
y = np.array ([3,8,1,10])
plt.plot(x,y,'o',color='r')
plt.xlabel('x axis points')
plt.ylabel('y axis points')
plt.title('narmal plot')
plt.show()
```

```python
x = np.array ([1,2,6,8])
y = np.array ([3,8,1,10])
plt.plot(x,y,color='y')
plt.xlabel('x axis points')
plt.ylabel('y axis points')
plt.title('narmal plot')
plt.show()
```

#  markers 

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = '*',ms= 10,mfc='r')
plt.show()
```

```python
y = np.array([3,8,1,10])
plt.plot(y,linestyle = 'dotted',color= 'g')
plt.show()
```
